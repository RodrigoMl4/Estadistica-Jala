## Actividad #2

- Calcular la media, rango, y desviación estándar de la variable 'price_usd' en la base de datos propuesta. 

```python
# Leer el archivo CSV como dataframe
automobiles_df = pd.read_csv('DB/cars.csv')
```

```python
automobiles_df.head()
```

```python
# Calcular la mediana de la columna 'price_usd'
media_price_usd = automobiles_df['price_usd'].mean()
```

```python
# Ordenar los valores de 'price_usd' en orden ascendente
automobiles_df_sorted = automobiles_df.sort_values(by='price_usd')
```

```python
# Calcular el rango de la columna 'price_usd'
rango_price_usd = automobiles_df_sorted['price_usd'].iloc[-1] - automobiles_df_sorted['price_usd'].iloc[0]
```

```python
# Calcular la desviación estándar de la columna 'price_usd'
desviacion_estandar_price_usd = automobiles_df['price_usd'].std()
```

```python
# Imprimir los resultados
print("La media de la columna 'price_usd' es:", media_price_usd)
print("El rango de la columna 'price_usd' es:", rango_price_usd)
print("La desviación estándar de la columna 'price_usd' es:", desviacion_estandar_price_usd)
```

## Actividad #4

- Calcular el primer, Segundo, tercer cuartile de la variable  'price_usd' en la base de datos. Discutir los resultados.

```python
# Leer el archivo CSV como dataframe
automobiles_df = pd.read_csv('DB/cars.csv')
```

```python
automobiles_df.head()
```

```python
# Calcular los cuartiles de la variable 'price_usd'
cuartil_1 = automobiles_df['price_usd'].quantile(0.25)
cuartil_2 = automobiles_df['price_usd'].quantile(0.50) 
cuartil_3 = automobiles_df['price_usd'].quantile(0.75)
```

```python
# Imprimir los resultados
print("Primer cuartil (Q1):", cuartil_1)
print("Segundo cuartil (Q2 o mediana):", cuartil_2)
print("Tercer cuartil (Q3):", cuartil_3)
```

### Conclusiones Ejercicio 4

### Primer cuartil (Q1):
El valor del primer cuartil (Q1), que es 2100.0 USD en tu conjunto de datos, indica que al menos el 25% de los precios de los automóviles son relativamente bajos, con un valor igual o inferior a este punto. 

### Segundo cuartil (Q2, Mediana):
El valor del segundo cuartil (Q2), también conocido como la mediana, es 4800.0 USD en tu conjunto de datos. Esto significa que el 50% de los precios de los automóviles son iguales o inferiores a este valor, y el otro 50% son iguales o superiores a este punto.

### Tercer cuartil (Q3):
El valor del tercer cuartil (Q3), que es 8990.0 USD en tu conjunto de datos, indica que al menos el 75% de los precios de los automóviles son iguales o inferiores a este punto.

```python
import matplotlib.pyplot as plt

# Datos de ritmo cardíaco
ritmo = [88, 80, 76, 72, 68, 56, 64, 60, 64, 68, 64, 68, 72, 76, 80, 84, 68, 80, 76, 72, 84, 80, 72, 76]

# Histograma
plt.hist(ritmo, bins=8, edgecolor='black')
plt.xlabel('Ritmo Cardíaco')
plt.ylabel('Frecuencia')
plt.title('Histograma de Ritmo Cardíaco')
plt.show()

# Diagrama de tallos y hojas
def stem_and_leaf(data):
    data.sort()
    stems = [int(x // 10) for x in data]
    leaves = [int(x % 10) for x in data]
    for stem, leaf in zip(stems, leaves):
        print(f"{stem} | {leaf}")

print("Diagrama de Tallos y Hojas:")
stem_and_leaf(ritmo)

# Diagrama de puntos
plt.plot(ritmo, 'bo', markersize=8)
plt.xlabel('Índice de Observación')
plt.ylabel('Ritmo Cardíaco')
plt.title('Diagrama de Puntos de Ritmo Cardíaco')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DB/study_performance.csv')
```

```python
df.describe()
```

```python
#Quantiles
quantiles = np.quantile(df['math_score'], [0, 0.25, 0.50, 0.75, 1])
```

```python
# Definir los porcentajes de los quantiles
percentiles = [0, 25, 50, 75, 100]
```

```python
print(quantiles)
```

```python
# Gráfico de barras para mostrar los quantiles
plt.bar(percentiles, quantiles)

# Etiquetas y título del gráfico
plt.xlabel('Percentiles')
plt.ylabel('Valor')
plt.title('Quantiles de la variable math_score')

# Mostrar la gráfica
plt.show()
```

```python
#Diagrama de cajas
plt.figure(figsize = (10,4))  # Crear figura e indicar tamañao
plt.boxplot(df['math_score'], vert = False) # crear el diagrama boxplot en forma horizontal
plt.title('Diagrama de cajas para la puntuacion de matematicas') # dar un titulo al diagrama
plt.ylabel('Puntuacion') # etiquetas en el eje y
# Mostrar la gráfica
plt.show()
```

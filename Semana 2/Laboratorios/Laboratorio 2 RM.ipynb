# Actividad #1 

Diseñar un gráfico de barras y un gráfico circular en los que represente la información para la variable color basándose en los datos proporcionados. ¿Cuál de los dos gráficos ilustra mejor la información?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
automobiles_df = pd.read_csv('DB/cars.csv')

color_counts = automobiles_df['color'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(color_counts.index, color_counts.values)
plt.xlabel('Color')
plt.ylabel('Count')
plt.title('Color Distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(color_counts.values, labels=color_counts.index, autopct='%1.1f%%')
plt.title('Color Distribution')
3plt.show()
```

# Conclusion
## Gráfico de barras

El gráfico de barras que generamos anteriormente ilustra de manera clara y efectiva la distribución de frecuencias de los diferentes colores. Las barras verticales permiten comparar fácilmente las cantidades de cada categoría de color. Este tipo de gráfico es especialmente útil cuando hay muchas categorías distintas, como es el caso en este conjunto de datos.

## Gráfico circular

Por otro lado, el gráfico circular (pie chart) también puede ser una opción para visualizar la distribución de la variable "color". Este tipo de gráfico divide un círculo en secciones proporcionales al porcentaje o frecuencia de cada categoría. Sin embargo, los gráficos circulares tienden a ser menos efectivos cuando hay muchas categorías diferentes, ya que puede ser difícil distinguir visualmente las secciones más pequeñas.

En cuanto a cuál de los dos gráficos ilustra mejor la información en este caso específico, considero que el gráfico de barras es la mejor opción. Las razones principales son:

1. Claridad visual: Con tantas categorías de colores distintos, el gráfico de barras permite comparar fácilmente las frecuencias y distinguir las diferencias entre las categorías más representadas y las menos representadas.

2. Precisión en la lectura de valores: En un gráfico de barras, es más sencillo estimar y comparar los valores exactos o rangos de frecuencia de cada categoría, mientras que en un gráfico circular puede resultar más complicado, especialmente para las secciones más pequeñas.

3. Escalabilidad: Si se agregaran más categorías de colores en el futuro, el gráfico de barras seguiría siendo una representación clara y efectiva, mientras que un gráfico circular podría volverse aún más difícil de interpretar.


# Actividad #2

Calcula una tabla de frecuencias para 'price_usd' en el conjunto de datos proporcionado, discute los resultados y preséntalos en un gráfico (por ejemplo, gráfico de barras, histograma). Tome como límite inferior del primer intervalo el primer cuartil Q1 y como limite superior del último intervalo el tercer cuartil Q3 el conjunto de datos proporcionado. Considere un número de clases apropiado escogiendo un valor entre 5 y 20.



```python
Q1 = automobiles_df['price_usd'].quantile(0.25)
Q3 = automobiles_df['price_usd'].quantile(0.75)

num_classes = 10 

freq_table = pd.cut(automobiles_df['price_usd'], bins=num_classes, 
                    include_lowest=True, labels=False)

freq_counts = freq_table.value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.bar(range(1, num_classes + 1), freq_counts)
plt.xlabel('Rango de Precios (USD)')
plt.ylabel('Frecuencia')
plt.title('Tabla de Frecuencias de Precios (USD)')

bin_width = (Q3 - Q1) / num_classes
tick_labels = [f'[{Q1 + i * bin_width:.2f}, {Q1 + (i+1) * bin_width:.2f})' for i in range(num_classes)]
plt.xticks(range(1, num_classes + 1), tick_labels, rotation=45)

plt.show()
```

# Conclusiones
Distribución de los datos: El histograma muestra una distribución sesgada hacia la derecha, lo que indica que la mayoría de los precios de los automóviles se concentran en el extremo inferior del rango de precios. Esto es evidente por la altura de la primera barra en comparación con las demás.
Rango de precios más común: El rango de precios más común para los automóviles es de $2100 a aproximadamente $2789, ya que esta es la barra más alta del histograma.
Disminución de la frecuencia: A medida que aumenta el rango de precios, la frecuencia de los automóviles dentro de esos rangos disminuye significativamente. Esto sugiere que hay menos automóviles disponibles a precios más altos.

# Actividad 3
Se ha realizado una encuesta para medir la satisfacción con una aplicación móvil. A cada usuario se le pidió que calificara la aplicación en una escala del 0 al 5, donde 5 representa la calificación más alta, indicando que la aplicación funciona excelente, y 0 denota que la aplicación no funciona bien en absoluto. Los resultados de 15 usuarios se muestran a continuación:

5, 5, 2, 4, 1, 3, 4, 0, 5, 2, 4, 1, 3, 2, 5

Realiza un análisis de las calificaciones obtenidas en esta encuesta y genera una tabla de frecuencias que muestre cuántos usuarios otorgaron cada calificación.


```python
calificaciones = [5, 5, 2, 4, 1, 3, 4, 0, 5, 2, 4, 1, 3, 2, 5]

arreglo_calificaciones = np.array(calificaciones)

valores_unicos, frecuencias = np.unique(arreglo_calificaciones, return_counts=True)

plt.bar(valores_unicos, frecuencias)

plt.xlabel("Calificación")
plt.ylabel("Frecuencia")

plt.title("Tabla de Frecuencias de Calificaciones de la Aplicación Móvil")

plt.show()
```

# Conclusiones

## Calificación más común

La calificación “5” fue la más comúnmente otorgada por los usuarios. Esto sugiere que la mayoría de los usuarios estaban satisfechos con la aplicación y la calificaron positivamente.

## Calificación más baja

La calificación “0” fue la menos común. Esto indica que muy pocos usuarios dieron la calificación más baja posible, lo que es una buena señal para la aplicación.

## Resolver ejercicio 5 de  online book in la página 159.

Una muestra de 30 puntuaciones de distancia medidas en yardas tiene una media de 10, una varianza de 9 y una desviación estándar de 3. (a) Quieres convertir todas tus distancias de yardas a pies, por lo que multiplicas cada puntuación de la muestra por 3. ¿Cuáles son la nueva media, varianza y desviación estándar? (b) Luego decides que solo quieres mirar la distancia más allá de un cierto punto. Por lo tanto, después de multiplicar las puntuaciones originales por 3, decides restar 4 pies a cada una de las puntuaciones. Ahora, ¿cuáles son la nueva media, varianza y desviación estándar?

# Solución

## Datos Iniciales

- Número de muestras (n) = 30 
- Media (mean) = 10 yardas
- Varianza (variance) = 9 yardas²
- Desviación estándar (nsd) = 3 yardas

## a) Convertir yardas a pies

Para convertir de yardas a pies, multiplicamos cada puntuación en la muestra por 3. 

- **Nueva media** = 3 * 10 = **30 pies**
- **Nueva varianza** = 9 * 3² = **81 pies²**
- **Nueva desviación estándar** = √81 = **9 pies**

## b) Restar una constante a los datos

Después de multiplicar las puntuaciones originales por 3, decides restar 4 pies a cada una de las puntuaciones.

- **Nueva media** = 30 pies - 4 = **26 pies**
- La varianza y la desviación estándar no cambian cuando se suma o resta una constante a los datos. Por lo tanto, la **varianza** sigue siendo **81 pies²** y la **desviación estándar** sigue siendo **9 pies**.


# Demostracion con Python

```python
import numpy as np
import pandas as pd

# Establecer una semilla para reproducibilidad
np.random.seed(0)

# Generar una muestra aleatoria de 30 distancias con media 10 y desviación estándar 3
distancias = np.random.normal(10, 3, 30)

# Crear un DataFrame de pandas
df = pd.DataFrame(distancias, columns=['Distancias en yardas'])

# Convertir yardas a pies
df['Distancias en pies'] = df['Distancias en yardas'] * 3

# Restar 4 pies a cada distancia
df['Distancias ajustadas'] = df['Distancias en pies'] - 4

# Calcular las medidas para cada conjunto de distancias
medidas = pd.DataFrame({
    'Media': df.mean(),
    'Varianza': df.var(),
    'Desviación estándar': df.std()
})

print(medidas)
```

# Justificacion con Python
Como podemos ver en el anterior codigo la varianza y la desviacion estandar no varian y solo cambia la media que se resta por la constante que en este caso es 4.
